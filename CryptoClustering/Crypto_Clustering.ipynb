```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
#hiding all warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
#checking total rows and columns
df_market_data.shape
```

```python

# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file

df_market_scaled = StandardScaler().fit_transform(df_market_data[["price_change_percentage_24h", "price_change_percentage_7d", "price_change_percentage_14d","price_change_percentage_30d","price_change_percentage_60d","price_change_percentage_200d","price_change_percentage_1y"]])
df_market_scaled 
```

```python
# Create a DataFrame with the scaled datadf_market_data[["price_change_percentage_24h", "price_change_percentage_7d", "price_change_percentage_14d","price_change_percentage_30d","price_change_percentage_60d","price_change_percentage_200d","price_change_percentage_1y"]])
df_market_scaled_df=pd.DataFrame(df_market_scaled, columns=["price_change_percentage_24h", "price_change_percentage_7d", "price_change_percentage_14d","price_change_percentage_30d","price_change_percentage_60d","price_change_percentage_200d","price_change_percentage_1y"])


# Copy the crypto names from the original data
df_market_scaled_df["coin_id"] = df_market_data.index
# Set the coinid column as index
df_market_scaled_df.set_index("coin_id", inplace =True)

# Display sample data
df_market_scaled_df.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 11))
```

```python
#Create an empty list to store the inertia values
inertia=[]

#Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for i in k_values:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(df_market_data)
    inertia.append(k_model.inertia_)
```

```python
#Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k_values, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)
df_elbow.head()
```

```python
#Plot a line chart with all the inertia values computed with 
#the different values of k to visually identify the optimal value for k.
df_elbow.hvplot.line(
    x="k", 
    y= "inertia", 
    title="Elbow Curve", 
    xticks=k_values
)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** The best value for 'k' is 4 as here inertia starts to decrease at a slower rate.

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
k_model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the scaled data
k_model.fit(df_market_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
predictions =k_model.predict(df_market_scaled)

# Print the resulting array of cluster values.
print(predictions)
```

```python
# Create a copy of the DataFrame
df_market_predictions= df_market_data.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
df_market_predictions['predicted_cluster'] = predictions 

# Display sample data
df_market_predictions.head()
```

```python
from sklearn.metrics import  silhouette_score
silhouette_score(df_market_data, k_model.labels_)
```

#a negative silhouette score indicates poor clustering performance, the clustering algorithm may not have effectively separated the data into distinct clusters. 

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
hover_cols = ["coin_id"]
cluster_colors = ['red', 'green', 'blue'] 
df_market_predictions.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="predicted_cluster",
    color=cluster_colors,
    hover_cols = hover_cols
)
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
df_market_scaled_pca = StandardScaler().fit_transform(df_market_data[["price_change_percentage_24h", "price_change_percentage_7d", "price_change_percentage_14d","price_change_percentage_30d","price_change_percentage_60d","price_change_percentage_200d","price_change_percentage_1y"]])
df_market_scaled_pca
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
market_data_pca = pca.fit_transform(df_market_scaled_pca)
# View the first five rows of the DataFrame. 
market_data_pca[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

```python
# Total explained variance ratio
pca_sum_explained_variance_ratio = sum(pca.explained_variance_ratio_)
print("Total Explained Variance Ratio:", pca_sum_explained_variance_ratio)
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** Total explained variance = 9.76037313e-01 + 02.30282949e-02+ 7.48308214e-04
The total explained variance of the three principal components is 0.8.

```python
# Create a new DataFrame with the PCA data.

# Creating a DataFrame with the PCA data
df_market_data_pca = pd.DataFrame(
    market_data_pca,
    columns=["PCA1", "PCA2","PCA3"]
    
)
# Copy the crypto names from the original data
df_market_data_pca["coin_id"] = df_market_data.index

# Set the coinid column as index
df_market_data_pca.set_index("coin_id", inplace =True)

# Display sample data
df_market_data_pca.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
#Create a list with the number of k-values from 1 to 11
k = list(range(1, 11))
```

```python
#Create an empty list to store the inertia values
inertia = []
#Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(df_market_data_pca)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data_pca = {"k": k, "inertia": inertia}
df_elbow_pca = pd.DataFrame(elbow_data_pca)

# Create a DataFrame with the data to plot the Elbow curve
df_elbow_pca.head()
```

```python
#Plot a line chart with all the inertia values computed with 
#the different values of k to visually identify the optimal value for k.
df_elbow_pca.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** the best value for `k` when using the PCA data is 5, as elbow point is 4, and from 5 it decresing.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:**  Yes, it is different from the best k value found using the original data.

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
K_model = KMeans(n_clusters=5, random_state=1)
```

```python
# Fit the K-Means model using the PCA data
K_model.fit(df_market_data_pca)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
k_3 = K_model.predict(df_market_data_pca)
# Print the resulting array of cluster values.
print(k_3)
```

```python
# Create a copy of the DataFrame with the PCA data
df_market_predictions_pca= df_market_data_pca.copy()

# Add a new column to the DataFrame with the predicted clusters

df_market_predictions_pca["predicted_clusters"] = k_3
# Display sample data
df_market_predictions_pca.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
hover_cols = ["coin_id"]
cluster_colors = ['red', 'green', 'blue'] 
df_market_predictions_pca.hvplot.scatter(
    x="PCA1",
    y="PCA2",
    by="predicted_clusters",
    color=cluster_colors,
    hover_cols = hover_cols
)
```
